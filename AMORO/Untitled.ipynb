```python
    def trajectory_Cartesian(start=np.array([0.09, 0.06796322]), end=np.array([0, 0.1]), duration = 2, ts = 0.01):
        """

        :param start:  start position
        :param end:  end position
        :param duration: time duration from 0 to duration
        :param ts: sampling time
        :return: x,y : position
                xd, yd: velocity
                xdd, ydd : acceleration
        """
        xa = start[0]
        xb = end[0]

        ya = start[1]
        yb = end[1]

        deltax = xb - xa
        detlay = yb - ya

        length = int(duration/ts) + 1

        x = np.zeros((length))
        y = np.zeros((length))
        xd = np.zeros((length))
        yd = np.zeros((length))
        xdd = np.zeros((length))
        ydd = np.zeros((length))


        t = 0
        for i in range(0, length):
            s = 10 * (t/duration)**3 - 15*(t/duration)**4 + 6*(t/duration)**5
            sd = 30/(duration)**3 * (t)**2 - 60/(duration)**4 *(t)**3 + 30/(duration)**5 * (t)**4
            sdd = 60/(duration)**3 * (t) - 180/(duration)**4 * (t) ** 2 + 120/(duration)**5 * (t) ** 3

            x[i] = xa + s * deltax
            y[i] = ya + s * detlay
            xd[i] = sd * deltax
            yd[i] = sd * detlay
            xdd[i] = sdd * deltax
            ydd[i] = sdd * detlay

            t = t+ts
        # plt.plot(x, y)
        # plt.plot(xd, yd)
        # plt.plot(xdd, ydd)
        # plt.show()
        return x, y, xd, yd, xdd, ydd

    # Trajectory for the biglide model
    x, y, xd, yd, xdd, ydd = trajectory_Cartesian(start=np.array([0, 0]), end=np.array([0, 2]))
```
